# **Decision Tree**

---



A Decision Tree for Best Performance

```python
table_and_treenum = {}
dt_model_perf = {}

for i in range(11):
  table = ds_names[i]
  X = table.drop(['Winner', 'R_fighter', 'B_fighter', 'weight_class'], axis=1)
  y = table['Winner']

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5)

  # determine the performance measure under each value of max depth for the dataset
  dt_perform_measure = {}

  for k in range(1, 50):

    # fit the single decision tree model with changing max depth
    dt = DecisionTreeClassifier(max_depth = k)
    dt.fit(X_train, y_train)

    # get the predictions
    predictions = dt.predict(X_test)

    # define the f1 score as the performance measure for each model
    tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
    f1_score = 2 * ((tp / (tp + fp)) * (tp / (tp + fn))) / ((tp / (tp + fp)) + (tp / (tp + fn)))

    # accumulate the result of performance measures
    dt_perform_measure[f1_score] = k

  best_performance = max(dt_perform_measure)
  dt_model_perf[i] = best_performance
  # find the depth of tree that gives the best performance
  depth = dt_perform_measure[max(dt_perform_measure)]
  # record it into a variable along with which table it belongs to
  table_and_treenum[i] = depth
```

```python
table_and_treenum
```

Final Result of DT

```python
final_dt_models = {}

# for each dataset
for i in table_and_treenum:

  # get the table name with their associated tree depth parameter from the previous dictionary
  table = ds_names[i]
  param = table_and_treenum[i]

  # define x and y variabl  es
  X = table.drop(['Winner', 'R_fighter', 'B_fighter', 'weight_class'], axis=1)
  y = table['Winner']

  # create and fit the model
  dt = DecisionTreeClassifier(max_depth = param)
  dt.fit(X, y)

  # record the weight class and the associated model
  final_dt_models[weight_classes[i]] = dt

  # print out final results
  print(weight_classes[i])
  fig = plt.figure(figsize=(30,30))
  cnames = [f"{weight_classes[i]}: Blue",f"{weight_classes[i]}: Red"]
  tree.plot_tree(dt, fontsize=12, feature_names=X.columns, class_names=cnames, filled=True);
```

```python
# tree_num_range = range(1, 500)
# max_features_range = range(1, 25)

tree_num_range = [5, 50, 100, 200, 300, 400, 500, 600, 700]
max_features_range = [1, 5, 8, 12, 15, 17, 19, 21, 25]
```

```python
X = df.drop(['Winner', 'R_fighter', 'B_fighter', 'weight_class'], axis=1)
y = df['Winner']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5, stratify = y)
```

```python
ploting_ds = pd.DataFrame()

performance = []

for i in tree_num_range:

  rf = RandomForestClassifier(n_estimators=i, criterion='gini')
  rf.fit(X_train, y_train)
  predictions = rf.predict(X_test)

  # define the f1 score as the performance measure for each model
  tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
  f1_score = 2 * ((tp / (tp + fp)) * (tp / (tp + fn))) / ((tp / (tp + fp)) + (tp / (tp + fn)))
  performance += [f1_score]

ploting_ds['Tree numbers'] = tree_num_range
ploting_ds['Performance_tree_num'] = performance

ploting_ds.plot.line('Tree numbers', 'Performance_tree_num')
```

```python
number_of_t = tree_num_range[performance.index(max(performance))]
```

```python
rf_ds_params = {}
rf_model_perf = {}

for i in range(11):

  table = ds_names[i]
  X = table.drop(['Winner', 'R_fighter', 'B_fighter', 'weight_class'], axis=1)
  y = table['Winner']

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5)

  rf_outcome_list = {}

  for tree_num in [200, 250, 300, 350, 400, 450, 500]:

    for feature_num in range(1, 5):

      rf = RandomForestClassifier(n_estimators=tree_num, max_features=feature_num, random_state=0, criterion='gini')

      rf.fit(X_train, y_train)

      predictions = rf.predict(X_test)

      # define the f1 score as the performance measure for each model
      tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
      f1_score = 2 * ((tp / (tp + fp)) * (tp / (tp + fn))) / ((tp / (tp + fp)) + (tp / (tp + fn)))

      rf_outcome_list[f1_score] = [tree_num, feature_num]

  best_performance = max(rf_outcome_list)
  rf_model_perf[i] = best_performance

  rf_params = rf_outcome_list[max(rf_outcome_list)]
  rf_ds_params[i] = rf_params


final_rf_models = {}

# for each dataset
for i in rf_ds_params:

  # get the table name with their associated tree depth parameter from the previous dictionary
  table = ds_names[i]
  param_tree_num = rf_ds_params[i][0]
  param_feature_num = rf_ds_params[i][1]

  # define x and y variables
  X = table.drop(['Winner', 'R_fighter', 'B_fighter', 'weight_class'], axis=1)
  y = table['Winner']

  # create and fit the model
  rm = RandomForestClassifier(n_estimators=param_tree_num, max_features=param_feature_num, random_state=0, criterion='gini')
  rm.fit(X, y)

  # record the weight class and the associated model
  final_rf_models[weight_classes[i]] = rm

final_rf_models
```

# **K Nearest Neighbours**

---




```python
table_and_kneighs = {}
knn_model_perf = {}

for i in range(11):
  table = ds_names[i]
  X = table.drop(['Winner', 'R_fighter', 'B_fighter', 'weight_class'], axis=1)
  y = table['Winner']

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2, stratify=y)

  # scale the dataset
  scaler = preprocessing.StandardScaler()
  scaler.fit(X_train)
  X_train_scaled = scaler.transform(X_train)
  X_test_scaled = scaler.transform(X_test)

  # determine the performance measure under each number of neighbours for the dataset
  knn_perform_measure = {}

  for k in range(1, int(len(table)*0.25)):

    # fit the knn model to the scaled dataset
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)

    # get the predictions
    predictions = knn.predict(X_test_scaled)

    # define the performance measures of each model
    tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
    f1_score = 2 * ((tp / (tp + fp)) * (tp / (tp + fn))) / ((tp / (tp + fp)) + (tp / (tp + fn)))

    # accumulate the result of performance measures
    knn_perform_measure[f1_score] = k

  best_performance = max(knn_perform_measure)
  knn_model_perf[i] = best_performance

  # find the number of neighbours that gives the best performance
  neighbours = knn_perform_measure[max(knn_perform_measure)]
  # record it into a variable along with which table it belongs to
  table_and_kneighs[i] = neighbours
```

```python
table_and_kneighs
```

```python
final_knn_models = {}

# for each dataset
for i in table_and_treenum:

  # get the table name with their associated tree depth parameter from the previous dictionary
  table = ds_names[i]
  param_neighbour_num = table_and_kneighs[i]

  # define x and y variables
  X = table.drop(['Winner', 'R_fighter', 'B_fighter', 'weight_class'], axis=1)
  y = table['Winner']

  # create and fit the model
  knn = KNeighborsClassifier(n_neighbors=param_neighbour_num)
  knn.fit(X, y)

  # record the weight class and the associated model
  final_knn_models[weight_classes[i]] = knn
```
